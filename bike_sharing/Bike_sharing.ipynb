# QD Test Model

```python
y=df['casual']
X=df.drop(['casual', 'yr'], axis='columns')
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y,test_size= 0.2)
```

```python
X_test
```

```python
ct=make_column_transformer((StandardScaler(), X.select_dtypes(include='number').columns),
                           remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pl=make_pipeline(ct, lr)
```

```python
pl.fit(X_train,y_train)
```

```python
pl.predict(X_test.loc[3397:3397, :])
```

```python
y.loc[3397]
```

```python
cross_val_score(pl, X, y, cv=20).mean()
```
